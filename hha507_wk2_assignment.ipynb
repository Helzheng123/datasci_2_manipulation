### 1. Data Cleaning with Pandas

#### Loading Data

Load the dataset *WK2/data/healthcare_data_cleaning.csv* in your Colab notebook

This can be found in datasets/healthcare_data_cleaning.csv

```python
data = pd.read_csv('https://raw.githubusercontent.com/Helzheng123/datasci_2_manipulation/main/datasets/healthcare_data_cleaning.csv', na_values=['missing'])
data
```

Identify and handle missing values in the dataset.

```python
# This step is to identify and count how many missing values is in the dataset
missing_values = data.isnull()
total_missing_values = missing_values.sum().sum()
print('Total missing values: ', total_missing_values)
```

```python
# Using data.fillna to fill the missing values as NaN
missing_values = data.fillna(np.nan)
missing_values
```

#### Remove any duplicate rows and columns, if they exist.

```python
# Remove duplicate rows through drop_duplicates
data.drop_duplicates(inplace=True)
data
```

```python
# Removing any duplicate columns where loc selects all the rows and only the columns that aren't duplicated.
# Used this site https://saturncloud.io/blog/how-to-remove-duplicate-columns-from-pandasreadcsv/#:~:text=Columns%20from%20pandas.-,read_csv(),by%20specifying%20the%20axis%20parameter. Not sure if it is correct but it is still a learning process.
data = data.loc[:, ~data.columns.duplicated()]
data
```

#### Clean column names

```python
# Remove leading and trailing white space from a column
data.columns= data.columns.str.strip()
data
```

```python
# Function to remove white space and special characters from a value
def clean_column_names(data):
  def clean_name(value):
    cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', value)
    return cleaned_name.lower()
  data.columns = [clean_name(col) for col in data.columns]
  return data
# Apply the clean_value function to all columns
data = clean_column_names(data)
```

```python
# Display clean data
data
```

## 2. Data Transformation

```python
# Used np.where to select the column used
# Did data[] to create a column
# Used this site https://www.digitalocean.com/community/tutorials/python-numpy-where
# If need to drop the column, do data = data.drop('column name', axis=1)

data['patientisover65'] = np.where(data['patientage'] > 65, True, False)
data['heartratestatus'] = np.where((data['averageheartrate'] > 60) & (data['averageheartrate'] < 100), 'Normal', 'Abnormal')
data
```

```python
# Here is to showcase the 4 specific columns used for creating the two new columns
selected_data = ['patientage', 'patientisover65', 'averageheartrate', 'heartratestatus']
selected_columns = data[selected_data]
selected_columns
```

#### Aggregate data using groupby and compute summary statistics

```python
# This is a summary statistics that helps group the disease diagnosed column and shows the different statistics of each column.
data.groupby(['diseasediagnosed']).describe()
```

```python
# This is to show only the counts for each column
grouped_data = data.groupby('diseasediagnosed').agg({'count'})
grouped_data
```

```python
# This is to show the payment method count for each disease diagnosed category
ways_to_pay = data.groupby(['diseasediagnosed', 'paymentmethod']).size().reset_index(name='count')
ways_to_pay
```

#### Use pivot tables or cross-tabulations for multi-dimensional analysis.

```python
# Disease diagnosed and payment pivot table
ways_to_pay.pivot_table(index='diseasediagnosed', columns='paymentmethod', values= 'count')
```

```python
# Another pivot table for counts based on gender and disease diagnosed
data.pivot_table(values='patientage', index='diseasediagnosed', columns='gender', aggfunc='count')
```
